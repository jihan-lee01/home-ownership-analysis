<h1 style="font-family: 'Arial', sans-serif; color: #2E86C1; font-size: 30px;">I. Introduction and Motivation</h1>

The topic of homeownership holds immense significance in the context of the United States, reflecting not only economic stability but also broader societal patterns. In recent years, the dynamics of homeownership have evolved, influenced by  demographic, economic, and social factors have changed. This project investigates the determinants of homeownership rates within the United States. 

According to Illinois Wesleyan University, the impact of home-ownership determinants has fluctuated with the economy: “Marital status remains important to the homeownership decision, though its effect has weakened. The impact of the level of education is now just as important as the presence of children in a household” (IWU,1999). The context of this study spans various societal and economic dynamics in the U.S. housing market. Historically, homeownership has been a critical measure of economic stability and societal status. The landscape of homeownership has evolved over time, influenced by shifts in economic conditions, demographic trends, and policy changes​.


Understanding the dynamics of homeownership is vital for policymakers, economists, and the general public. It sheds light on economic disparities, social mobility, and the overall health of the real estate market. This research could inform policies aimed at addressing inequalities in housing and aiding those struggling to attain homeownership.

<h2 style="font-family: 'Arial', sans-serif; color: #2E86C1; font-size: 25px;">a. Literature Review</h2>

Various studies highlight the significance of factors like marital status, race, gender, and income in homeownership. For instance, married couples tend to have higher homeownership rates compared to unmarried individuals, illustrating the influence of combined income and stability.“Marital status remains important to the homeownership decision, though its effect has weakened. The impact of the level of education is now just as important as the presence of children in a household” (IWU,1999).

 Higher family incomes not only provide the means to cover initial homeownership costs but also improve credit conditions, making homeownership more accessible and financially attractive​ (Hood)​. Racial disparities in homeownership rates are evident, with white households owning homes at higher rates than black households, partly due to historical discrimination and differences in income. The role of gender is also significant, as male household heads often have more stable incomes, influencing their likelihood of homeownership. Furthermore, income level is a crucial determinant; higher income increases the likelihood of owning a home due to the ability to cover upfront costs and secure favorable mortgage terms (Carliner). According to the IWU research, important factors like marital status, education level, and the presence of children significantly influence homeownership. These elements' impact on homeownership decisions has evolved alongside societal and economic changes​​. According to Brookings analyst Jenny Schuetz, discrimination in housing has contributed to substantial wealth gaps between Black and white households. Black and Latino or Hispanic families are less likely to own their homes than white families with similar incomes. Older households, typically having more wealth and income certainty, are more likely to invest in homeownership. This trend correlates with their lower mobility and higher ability to diversify investments​​ (Hood).

```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from stargazer.stargazer import Stargazer
from IPython.core.display import HTML

# Import data
cps = pd.read_csv("/Users/jihan/Desktop/econ-320-lab-project/econ-320-lab-project/data/cps.csv")
```

<h1 style="font-family: 'Arial', sans-serif; color: #2E86C1; font-size: 25px;">a.Cleaning</h1>

```python
cps['REGION'] = cps['REGION'].replace({11:0, 12:0, 21:1, 22:1, 31:2, 32:2, 33:2, 41:3, 42:3}) # 0: Northeast Region, 1: Midwest Region, 2: South Region, 3: West Region
cps['OWNERSHP'] = cps['OWNERSHP'].replace({10:1, 21:0, 22:0}) # 0: Not Owned, 1: Owned
cps['SEX'] = cps['SEX'].replace({2:1, 1:0}) # 0: Male, 1: Female
cps['RACE'] = cps['RACE'].replace({100:0, 200:1, 300:2, 651:3, 652:3}) # 0: White, 1: Black, 2: American Indian/Aleut/Eskimo, 3: Asian or Pacific Islander, 4: Two or More Races
cps['RACE'] = cps['RACE'].astype(str)
cps.loc[cps['RACE'].str.match('8\d{2}'), 'RACE'] = 4
cps['MARST'] = cps['MARST'].replace({1:0, 2:0, 3:1, 4:1, 5:1, 6:1}) # 0: Married, 1: Single
cps.drop(cps[cps['UHRSWORKLY'] == 999].index, inplace=True) # Filter out NIU values
cps = cps[cps['RELATE'] == 101] # Filter out only Head/householder
cps = cps[cps['INCTOT'] > 0] # Filter out negative income values
cps['LOGINCTOT'] = np.log(cps['INCTOT']) # Add log-transformed income variable
# 0: None, preschool, or kindergarten, 1: Primary education (Grade 1-6), 2: Secondary education (Grade 7-12), 3: Some college/associate's degree, 4: Bachelor's degree, 5: Postgraduate education
cps['EDUC'] = cps['EDUC'].replace({2:0, 10:1, 20:1, 30:2, 40:2, 50:2, 60:2, 71:2, 73:2, 81:3, 91:3, 92:3, 111:4, 123:5, 124:5, 125:5}) 
# Separate 2007 and 2009 data
cps2007 = cps[cps['YEAR'] == 2007]
cps2009 = cps[cps['YEAR'] == 2009]
# Drop YEAR and RELATE variables
cps2007 = cps2007.drop(['YEAR', 'RELATE'], axis=1)
cps2009 = cps2009.drop(['YEAR', 'RELATE'], axis=1)
```

```python
# Functions for plots
def histogram(variable, title, xlabel, ax):
    ax.hist(variable, edgecolor='black')
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel('Frequency')

def piechart(variable, title, label, ax):
    labels = variable.map(label).value_counts().index
    sizes = variable.map(label).value_counts().values
    ax.pie(sizes, autopct='%1.2f%%', startangle=140)
    ax.legend(labels, title="Categories", loc="best")
    ax.set_title(title)
    ax.axis('equal')

def heatmap(corr, title):
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title(title)
    plt.show()
```

**Economic Hypotheses**

*Sex:* We think sex is a relevant variable to employ in our dataset as we believe men are more likely to own a house due to the gender wage gap.

*Race:* Based on our research, ethnicity and race play a factor in home-ownership due to discrimination in the home-buying process and the general wealth gap between people of color and white individuals.

*Employment Status:* Employment status affects income and credit status, as well as the turnover rate for mortgage and ownership applications.

*Education Level:* Education affects job attainment as well as future income. Education is also an indicator of background, where those who come from higher economic backgrounds are more likely to attend school or come from wealth, and therefore, more likely to own a home.

*Citizenship Status:* Non-citizens may encounter difficulties accessing mortgage loans, face varied down payment requirements, and confront challenges in building a credit history. Additionally, their income and employment opportunities may impact their ability to afford a home, and some government programs may not be accessible to non-citizens. Fear of deportation among undocumented immigrants can also discourage home ownership.

*Marital Status:* Based on our preliminary research, we observed that the highest share of home buyers in the United States in 2022 amongst all age groups was married couples. We believe this can be attributed to a higher income stability and combined incomes between the couple.

*Income:* We believe that income holds a significant role in determining the household status of a person. The higher the income, the larger one’s disposable income, given those who earn a higher wage are more likely to be able to afford a home.  

*Number of Family Members in Household:* We presume that the number of family members in a household will affect one’s household status. Given higher income stability of those who are married with children, we believe they are more likely to own a home.

*Disability:* Whether a person is disabled or not significantly determines the likelihood they own a home. From preliminary research, we noted that the lack of accessible and affordable housing is one of the largest obstacles people with disabilities face in obtaining housing. Given this, we presume that people with disabilities are less likely to own homes.

*Veteran Status:* Veteran Status is a pertinent variable in our dataset given the discrepancies between veterans and non-veterans in economic status. In our preliminary research, we observed that roughly ⅕ of the homeless population consists of veterans, which leads us to the belief that veterans are less likely to own homes than non-veterans.
